```python
#step  ______1 _______import lab


import pandas as pd#____________for making the dataframe
import numpy as np
import matplotlib.pyplot as plt#____________for ploting  the data
from sklearn.preprocessing import StandardScaler #____________for normal the data
from sklearn.preprocessing import MinMaxScaler #____________for normal the data
from sklearn.model_selection import train_test_split#____________for training  the data
import seaborn as  sns #____________for ploting and load   the data
```

```python
df=pd.read_csv("C:\\Users\\Rizwan\\Desktop\\train.csv")  #------step 2 load data
```

```python
df.head()
```

```python
pd.set_option("display.max_rows", None)  #------to check all cols and rows
pd.set_option("display.max_columns",None)
```

```python
df.head()
```

```python
#step___2 ___making new_df for input vals /independent
df1=df[["MSSubClass","LotFrontage","LotArea","SalePrice","GarageArea"]]
```

```python
df1.head(10)


#SalePrice__________is output varl so we drop it from train and put it in test
```

```python
#step_3__________check null vals and fill them
df1.isnull().sum()
```

```python
df_fill=df1.fillna(df1.mean()).round(2)

df_fill
```

```python
#step__4___________drop the output col and put it in new df
X=df_fill.drop("SalePrice",axis=1)
Y=df_fill["SalePrice"]
print("Shape of X", X.shape)
print("Shape of Y", Y.shape)
```

```python
#step_5_______making (x_train,xtest and y_train,y_test)vars

#____we split data into 20% in test and 80% in train 
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,train_size=0.8,random_state=50)
```

```python
print("Shape of X_tarin", x_train.shape)
print("Shape of X_test", x_test.shape)
print("Shape of Y_tarin", y_train.shape)
print("Shape of Y_test", y_test.shape)
```

```python
#step__6________fit the train data
sc=StandardScaler()
sc.fit(x_train)
```

```python
sc.mean_
```

```python
#check the statistic of train data  before feauter scalin/transform
x_train.describe()
```

```python
#step_7_________tranform the x_train and x_test data
x_train_sc_trf=sc.transform(x_train)
x_test_sc_trf=sc.transform(x_test)
```

```python
x_test_sc_trf
```

```python
x_train_sc_trf
```

```python
#step_8________making (x_train_sc_trf,x_test_sc_trf) dataframe
x_train_df=pd.DataFrame(x_train_sc_trf,columns=["MSSubClass","LotFrontage","LotArea","GarageArea"])
x_test_df=pd.DataFrame(x_test_sc_trf,columns=["MSSubClass","LotFrontage","LotArea","GarageArea"])
```
