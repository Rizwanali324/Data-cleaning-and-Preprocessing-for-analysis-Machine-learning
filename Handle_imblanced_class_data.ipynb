# Preprocessing

```python
df.drop(['fnlwgt', 'education','race'],axis=1,inplace=True)
```

```python
df.sample(9)
```

```python
# Map the values in the 'income' column to 0 and 1
income_mapping = {'<=50K': 0, '>50K': 1}
df['income'] = df['income'].map(income_mapping)
```

```python
df.sample(9)
```

```python
df.info()
```

```python
df.keys()
```

```python

df['income'] = df['income'].astype(str)
```

```python
df.info()
```

```python
# Replace unknown values with mode for string columns
string_columns = df.select_dtypes(include='object').columns
for column in string_columns:
    mode_value = df[column].mode().iloc[0]
    df[column] = df[column].replace(r'\?', mode_value, regex=True)

# Replace unknown values with mean for integer columns
integer_columns = df.select_dtypes(include='number').columns
for column in integer_columns:
    mean_value = df[column].mean()
    df[column] = df[column].fillna(mean_value)
```

```python

from imblearn.over_sampling import RandomOverSampler

# Assuming you have loaded your dataset into a pandas DataFrame called "df"

# Check the class distribution of "income"
class_distribution = df['income'].value_counts()

# Print the class distribution
print(class_distribution)

# Plot the class distribution before handling imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='income', data=df)
plt.title('Income Class Distribution (Before Handling Imbalance)')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()

# Handling imbalance using Random Over-Sampling
X = df.drop('income', axis=1)
y = df['income']

ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Create a new DataFrame with the resampled data
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['income'] = y_resampled

# Check the class distribution after handling imbalance
class_distribution_resampled = df_resampled['income'].value_counts()

# Print the class distribution after handling imbalance
print(class_distribution_resampled)

# Plot the class distribution after handling imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='income', data=df_resampled)
plt.title('Income Class Distribution (After Handling Imbalance)')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

imblanced data can lead to low accuracy in classification models
