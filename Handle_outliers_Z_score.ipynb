```python
import pandas as pd

import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the number of rows and columns for the DataFrame
num_rows = 1000
num_columns = 10

# Set the outlier parameters
outlier_fraction = 0.05  # Proportion of outliers in each column
outlier_range = (-10, 10)  # Range within which outliers can occur

# Create a dictionary to store the column data
data = {}

# Generate random values from a normal distribution for each column
for i in range(num_columns):
    column_name = f'Column_{i+1}'
    
    # Generate normally distributed values
    values = np.random.normal(0, 1, num_rows)
    
    # Introduce outliers
    num_outliers = int(num_rows * outlier_fraction)
    outlier_indices = np.random.choice(num_rows, num_outliers, replace=False)
    values[outlier_indices] = np.random.uniform(outlier_range[0], outlier_range[1], num_outliers)
    
    data[column_name] = values
df = pd.DataFrame(data)
df
```

```python
# Create a figure and axes for the box plots
fig, axes = plt.subplots(nrows=1, ncols=len(df.columns), figsize=(12, 4))

# Iterate over each column and create a box plot
for i, column in enumerate(df.columns):
    axes[i].boxplot(df[column])
    axes[i].set_title(column)

# Set the y-axis label for all subplots
fig.text(0.04, 0.5, 'Values', va='center', rotation='vertical')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()

```

# Removing the outliers

```python


# Iterate over each column in the DataFrame
for column in df.columns:
    # Get the top and bottom 5 percentile values
    top_threshold = np.percentile(df[column], 95)
    bottom_threshold = np.percentile(df[column], 5)
    
    # Find the outliers based on the thresholds
    column_outliers = df[(df[column] > top_threshold) | (df[column] < bottom_threshold)]
    
    # Print the outliers for the column
    print("Outliers in column", column, ":")
    print(column_outliers)
    print()
```

```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Calculate the mean and standard deviation for the column
    mean = df[column].mean()
    std = df[column].std()
    
    # Define the upper and lower thresholds
    upper_threshold = mean + 3 * std
    lower_threshold = mean - 3 * std
    
    # Remove the outliers from the column
    df = df[(df[column] <= upper_threshold) & (df[column] >= lower_threshold)]

# Create a new dataset 'final_df' without the outliers
final_df = df.copy()

# Print the final dataset without outliers
print("Final dataset without outliers:")
final_df
```

```python
for column in final_df.columns:
    plt.figure()
    sns.kdeplot(final_df[column], shade=True)
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title('Distribution of ' + column)
    plt.show()
```

# capping

```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Calculate the mean and standard deviation for the column
    mean = df[column].mean()
    std = df[column].std()
    
    # Define the upper and lower thresholds
    upper_threshold = mean + 3 * std
    lower_threshold = mean - 3 * std
    
    # Identify the extreme values for the column
    extreme_values = df[(df[column] > upper_threshold) | (df[column] < lower_threshold)][column]
    
    # Determine the number of extreme values exceeding the limit
    num_extreme_values = len(extreme_values)
    
    # If there are more than 5 extreme values, cap them to the respective threshold
    if num_extreme_values > 5:
        capped_extreme_values = extreme_values.copy()
        
        # Cap values above the upper threshold
        capped_extreme_values[capped_extreme_values > upper_threshold] = upper_threshold
        
        # Cap values below the lower threshold
        capped_extreme_values[capped_extreme_values < lower_threshold] = lower_threshold
        
        # Update the DataFrame with the capped values
        df.loc[extreme_values.index, column] = capped_extreme_values
        # Print the capped values for the column
    print("Capped values in column", column, ":")
    print(df[(df[column] == upper_threshold) | (df[column] == lower_threshold)][column])
    print()

# Print the final capped DataFrame
print("Final Capped DataFrame:")
print(df)
    
```

# for both zscore and iqr

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro

# Set random seed for reproducibility
np.random.seed(123)

# Generate data for normal distribution columns
normal_data = np.random.normal(loc=0, scale=1, size=(100, 6))

# Generate data for non-normal distribution columns
non_normal_data = np.random.exponential(scale=1, size=(100, 6))

# Combine normal and non-normal data
data = np.concatenate((normal_data, non_normal_data), axis=1)

# Create a DataFrame from the generated data
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L'])

# Check normality and apply z-score or IQR method accordingly
z_score_threshold = 3  # Threshold for z-score outliers
iqr_multiplier = 1.5  # Multiplier for IQR outliers

for col in df.columns:
    # Check normality using Shapiro-Wilk test
    _, p_value = shapiro(df[col].dropna())
    
    if p_value > 0.05:
        # Apply z-score method for columns following a normal distribution
        z_scores = (df[col] - df[col].mean()) / df[col].std()
        df[col] = np.where(abs(z_scores) > z_score_threshold, np.nan, df[col])
    else:
        # Apply IQR method for columns not following a normal distribution
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_multiplier * IQR
        upper_bound = Q3 + iqr_multiplier * IQR
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), np.nan, df[col])

# Fill outliers with nearest non-outlier values
df_capped = df.fillna(method='bfill').fillna(method='ffill')

# Create a new DataFrame with capped outliers
new_df = pd.DataFrame(df_capped, columns=df.columns)

# Print the new DataFrame\\
new_df
```
