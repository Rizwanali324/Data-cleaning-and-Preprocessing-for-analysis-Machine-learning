```python
#_____check duplicate values
df.duplicated(keep=False).sum()
```

```python
#____________alll duplicate values
df.loc[df.duplicated(keep=False)]
```

```python
#________By defualt remove all duplicate rows
df_drop_rows=df.drop_duplicates()
df_drop_rows##__________change in rows
```

```python
#To remove duplicates on specific column(s), use ``subset``.
df_drop_cols_dup=df.drop_duplicates(subset=["occupation"])
df_drop_cols_dup
```

```python
df_keep_first=df.drop_duplicates(keep="first")
df_keep_first
```

```python
df_keep_last=df.drop_duplicates(keep="last")
df_keep_last
```

```python
print("shape of original df : ",df.shape)
print("shape of drop_cols df : ",df_drop_cols_dup.shape)
print("shape of drop_rows df : ",df_drop_rows.shape)
print("shape of keep_first df : ",df_keep_first.shape)
print("shape of keep_lastl df : ",df_keep_last.shape)
```
