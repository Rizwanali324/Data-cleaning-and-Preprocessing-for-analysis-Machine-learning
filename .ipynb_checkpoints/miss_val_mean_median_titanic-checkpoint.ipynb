```python
miss_val_pers=df1.isnull().sum()/df1.shape[0]*100
```

```python
miss_val_pers
```

```python
miss_val_col=miss_val_pers[miss_val_pers>20].keys()
```

```python
df1_drop_col=df1.drop(columns=miss_val_col)
```

```python
df1_drop_col.shape
```

```python
df1_numeric=df1_drop_col.select_dtypes(include=['int64','float64'])
```

```python
df1_numeric.head(8)
```

```python
df1_numeric.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df1_numeric)
```

```python
df1_numeric[df1_numeric.isnull().any(axis=1)]
```

```python
df1_numeric.isnull().sum()
```

```python
df1_numeric.isnull().sum().sum()
```

```python
missing_num_val=[var for var in df1_numeric.columns if df1_numeric[var].isnull().sum()>0]
```

```python
missing_num_val
```

```python
plt.figure(figsize=(16,16))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df1_numeric[var], bins=20, kde_kws={'linewidth':5, 'color':'k'})
```

```python
df1_mean=df1_numeric.fillna(df1_numeric.mean())
df1_mean.
```

```python
df1_mean.isnull().sum().sum()
```

```python
df1_median=df1_numeric.fillna(df1_numeric.median())
sns.heatmap(df1_median)
```

```python
df1_median
```

```python
df1_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,16))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df1_median[var], bins=20, kde=True)
```

```python
plt.figure(figsize=(16,16))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df1_mean[var], bins=20,kde=True)
```
