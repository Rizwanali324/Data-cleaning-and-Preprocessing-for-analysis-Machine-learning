# Check nulls vals

```python
df.isnull().sum()
```

# taking persage of missing val

```python
miss_val_pers=df.isnull().sum()/df.shape[0]*100
```

```python
miss_val_pers
```

# pick the missing val cols

```python
miss_val_col=miss_val_pers[miss_val_pers>20].keys()
```

```python
miss_val_col
```

# drop missing val cols >20%

```python
df_drop_col=df.drop(columns=miss_val_col)
```

```python
df_drop_col.shape
```

# select num datas

```python
df_numeric=df_drop_col.select_dtypes(include=['int64','float64'])
```

```python
df_numeric.sample(7)
```

```python
df_numeric.head()
```

# shows missing values in graphs

```python
plt.figure(figsize=(20,20))
sns.heatmap(df_numeric)#plot ans see mising val after  drop
```

```python
df_numeric[df_numeric.isnull().any(axis=1)]#drop missing val about rows
```

```python
df_numeric.isnull().sum()
```

```python
missing_num_val=[var for var in df_numeric.columns if df_numeric[var].isnull().sum()>0]#store all missing val in var
```

```python
missing_num_val
```

# making baning

```python
plt.figure(figsize=(16,16))#check dist of missing data val
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df_numeric[var], bins=20, kde_kws={'linewidth':5, 'color':'k'})
```

```python
df_mean=df_numeric.fillna(df_numeric.mean())#fill missing val with mean of missing vals
df_mean
```

```python
df_mean.isnull().sum().sum()
```

Data clean with means

# Data clean with modes

```python
df_median=df_numeric.fillna(df_numeric.median())#ffill missing val with meading of missing vlas
```

```python
df_median.sample(6)
```

```python
df_median.isnull().sum().sum()#____________check nulss vals
```

# check the change in distrebution of data

```python
plt.figure(figsize=(16,16))#check dist of data after fill missing val with medien 
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df_median[var], bins=20, kde=True)
```

```python
plt.figure(figsize=(16,16))#check dist oof data after filling missing val with mean of missing val 
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df_mean[var], bins=20,kde=True,color="g")
```

```python
plt.figure(figsize=(16,16))#check dist oof data after filling missing val with mean of missing val 
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.histplot(df_mean[var], bins=20,kde=True,label="mean")
    sns.histplot(df_median[var], bins=20, kde=True,label="medien")
    
    sns.histplot(df_numeric[var], bins=20, kde_kws={'linewidth':5, 'color':'k'},label="Original")
    plt.legend()
```
